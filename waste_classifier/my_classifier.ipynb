```python
# import required libraries
import cvzone
import os
import cv2
from cvzone.ClassificationModule import Classifier
```

```python
# connect to camera
cap = cv2.VideoCapture(0)

# import trained model from teachable machine
classifier = Classifier('Model/keras_model.h5', 'Model/labels.txt')
```

```python
# Import bin images
imgBinsList = []
pathFolderBins = "Bins"
pathList = os.listdir(pathFolderBins)
pathList.pop(0)
pathList.sort()
print(pathList)
for path in pathList:
    imgBinsList.append(cv2.imread(os.path.join(pathFolderBins, path), cv2.IMREAD_UNCHANGED))
```

```python
# 0 = Organic
# 1 = Paper
# 2 = General waste
# 3 = Glass
# 4 = Plastic
 
classDic = {0: 5,
            1: 0,
            2: 4,
            3: 2,
            4: 1,
            5: 3,
           }
```

```python
# run and classify waste
while True:
    _, img = cap.read()
    imgResize = cv2.resize(img, (1240, 695))
    img_back = cv2.imread("background.png")
    
    
    predection = classifier.getPrediction(img)
    classID = predection[1]
    
    classIDBin = classDic[classID]
    img_back = cvzone.overlayPNG(img_back, imgBinsList[classIDBin], (1700, 390))
    
    img_back[515:515+695, 200:200+1240]= imgResize
    
    cv2.imshow("Output", img_back)
    cv2.waitKey(1)
```
